```python
# attempt to pivot the monthly information. Would love it formatted for YEAR and PERIOD columns

# only work with the Virginia data for the wrangling
# dfVAZillow2BedRm contains on the VA state zip codes
dfZ3BdRmMtroVA = dfZ3BdRmMtro[dfZ3BdRmMtro['StateName'].str.contains("VA")].head(5)

#use melt to change monthly dates into rows
dfZ3BdRmMtroVA = dfZ3BdRmMtroVA.melt(id_vars=['RegionID','SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'City', 'Metro', 'CountyName'], var_name='YYYY-MM-DD')
dfZ3BdRmMtroVA
```

### Need to create a column for yr, and period to match the HPI data set

```python
dfZ3BdRmMtroVA['yr'] = pd.DatetimeIndex(dfZ3BdRmMtroVA['YYYY-MM-DD']).year
dfZ3BdRmMtroVA['period'] = pd.DatetimeIndex(dfZ3BdRmMtroVA['YYYY-MM-DD']).month
```
