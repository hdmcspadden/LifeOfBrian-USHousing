```python
# website where the gross domestic product can be found in various forms (top level, by state, by industry, etc.)
# https://www.bea.gov/data/gdp
```

```python
list_df = tabula.read_pdf('GDP1995-2001.pdf', pages = "8", stream = True)
```

```python
df = list_df[0]
```

```python
df
```

```python
column_names = {'Unnamed: 0':'Area', 'Unnamed: 1':'1995', 'Unnamed: 2':'1996', 'Unnamed: 3':'1997', 'Unnamed: 4':'1998', 'Unnamed: 5':'1999', 
                'Unnamed: 6':'2000', 'Unnamed: 7':'2001', 'Percentage of U.S. total':'Percentage of U.S. total'}
summary = ['United States', 'New England', 'Mideast', 'Great Lakes',
           'Plains', 'Southeast', 'Southwest', 'Rocky Mountain', 'Far West']
```

```python
df = df.drop(df.index[0])
```

```python
GDP95_01df = df.rename(columns = column_names).reset_index(drop=True)
```

```python
GDP95_01df
```

```python
GDP95_01df['Area'] = GDP95_01df['Area'].astype('str')
```

```python
GDP95_01df
```

```python
# clean the white space and periods from the area column
# create a list to hold the cleaned area names
clean_areas = []
for i in GDP95_01df['Area'].index:
    if i == 0:
        print('Pass')
    else:
        area = GDP95_01df.loc[i, 'Area']
        #print(area)
        clean_areas.append(' '.join(re.sub("[\s\.*]"," ", area).split()))
        
# Loop through the list to add the cleaned areas back into the DataFrame using loc() method
for i in GDP95_01df['Area'].index:
    if i == 0:
        print('Pass')
        pass
    else:
        #print(i-1)
        GDP95_01df.loc[i , 'Cleaned Area'] = clean_areas[i-1]
    
```

```python
GDP95_01df
```

```python
for i in GDP95_01df['Cleaned Area'].index:
    if i == 0:
        print('Pass')
        pass
    else:
        if GDP95_01df.loc[i, 'Cleaned Area'] in summary:
            print(GDP95_01df.loc[i, 'Cleaned Area'] + ', Index = ' + str(i))
            #if GDP95_01df.loc[i, 'Cleaned Area'] == j:
            #    GDP95_01df = GDP95_01df.drop(GDP95_01df.index[i])
            
```

```python
# Add a categorical column that assigns a category to each state. Does not a category to the summary lines
for i in range(3,9):
    GDP95_01df.loc[i, 'Geo Loc'] = 'New England'
for i in range(10,16):
    GDP95_01df.loc[i, 'Geo Loc'] = 'Mideast'
for i in range(17,22):
    GDP95_01df.loc[i, 'Geo Loc'] = 'Great Lakes'
for i in range(23,30):
    GDP95_01df.loc[i, 'Geo Loc'] = 'Plains'
for i in range(31,43):
    GDP95_01df.loc[i, 'Geo Loc'] = 'Southeast'
for i in range(44,48):
    GDP95_01df.loc[i, 'Geo Loc'] = 'Southwest'
for i in range(49,54):
    GDP95_01df.loc[i, 'Geo Loc'] = 'Rocky Mountain'
for i in range(55,61):
    GDP95_01df.loc[i, 'Geo Loc'] = 'Far West'

    
GDP95_01df
```

```python
percOfTotal_2000 = []
percOfTotal_2001 = []
```

```python
for i in GDP95_01df['Percentage of U.S. total'].index:
    if i == 0:
        pass
    else:
        splitList = GDP95_01df.loc[i, 'Percentage of U.S. total'].split(" ")
        percOfTotal_2000.append(splitList[0])
        percOfTotal_2001.append(splitList[2])
```

```python
percOfTotal_2001
```

```python
for i in range(0, len(percOfTotal_2000)):
    GDP95_01df.loc[i+1,'percOfTotal_2000'] = percOfTotal_2000[i]
    
for i in range(0, len(percOfTotal_2001)):
    GDP95_01df.loc[i+1,'percOfTotal_2001'] = percOfTotal_2001[i]
```

```python
states = ["Alaska", "Alabama", "Arkansas", "Arizona", "California", "Colorado", "Connecticut", "District of Columbia",
          "Delaware", "Florida", "Georgia", "Hawaii", "Iowa", "Idaho", "Illinois", "Indiana", "Kansas", "Kentucky", 
          "Louisiana", "Massachusetts", "Maryland", "Maine", "Michigan", "Minnesota", "Missouri", "Mississippi", "Montana", 
          "North Carolina", "North Dakota", "Nebraska", "New Hampshire", "New Jersey", "New Mexico", "Nevada", "New York", 
          "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", 
          "Texas", "Utah", "Virginia", "Vermont", "Washington", "Wisconsin", "West Virginia", "Wyoming"]
GDPdf = pd.DataFrame()
count = 0
```

```python
for year in range(1995,2002):
    for state in states:
        GDPdf.loc[count, 'year'] = str(year)
        GDPdf.loc[count, 'state'] = state
        GDPdf.loc[count, 'GDP'] = GDP95_01df.loc[GDP95_01df['Cleaned Area'] == state, str(year)].item()
        GDPdf.loc[count, 'GDP_area'] = GDP95_01df.loc[GDP95_01df['Cleaned Area'] == state, 'Geo Loc'].item()
        count += 1

# source for .loc[].item() method: 
# https://stackoverflow.com/questions/36684013/extract-column-value-based-on-another-column-pandas-dataframe
```
