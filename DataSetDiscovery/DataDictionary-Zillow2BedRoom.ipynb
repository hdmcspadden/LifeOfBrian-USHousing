```python
# attempt to pivot the monthly information. Would love it formatted for YEAR and PERIOD columns

# only work with the Virginia data for the wrangling
#dfVAZillow2BedRm contains on the VA state zip codes
dfVAZillow2BedRm = dfZ2BdRmMtro[dfZ2BdRmMtro['StateName'].str.contains("VA")].head(5)

#use melt to change monthly dates into rows
dfVAZillow2BedRm = dfVAZillow2BedRm.melt(id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'City', 'Metro', 'CountyName'], var_name='YYYY-MM-DD')
dfVAZillow2BedRm

#dfZ2BdRmMtroPivot = dfZ2BdRmMtro.melt(id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'City', 'Metro', 'CountyName'], var_name='YYYY-MM-DD')
#dfZ2BdRmMtroPivot
```

Determine how many NaN's are in value

```python
print("number of NaN values for the column bedrooms :", dfVAZillow2BedRm['value'].isnull().sum())
```

We are only missing one value: Chesapeake City county for January 1996. Replace this NaN with the value for February 1996 for Chesapeake.

```python
dfFeb1996Value = dfVAZillow2BedRm[dfVAZillow2BedRm['CountyName'] == 'Chesapeake City']
valueFeb1996 = dfFeb1996Value['value'].iloc[1]

dfVAZillow2BedRm['value'] = np.where((dfVAZillow2BedRm.value.isnull()),valueFeb1996,dfVAZillow2BedRm.value)

print("number of NaN values for the column bedrooms :", dfVAZillow2BedRm['value'].isnull().sum())

dfVAZillow2BedRm.head(10)
```

### Need to create a column for yr, and period to match the HPI data set

```python
dfVAZillow2BedRm['yr'] = pd.DatetimeIndex(dfVAZillow2BedRm['YYYY-MM-DD']).year
dfVAZillow2BedRm['period'] = pd.DatetimeIndex(dfVAZillow2BedRm['YYYY-MM-DD']).month
```
